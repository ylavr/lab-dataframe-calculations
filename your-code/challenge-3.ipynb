## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

orders_grouped = orders.groupby(["CustomerID"]).agg({"amount_spent":"sum"})
```

```python
display(orders_grouped)
```

```python
quantile_95 = orders_grouped["amount_spent"].quantile(q=0.95)
quantile_75 = orders_grouped["amount_spent"].quantile(q=0.75)


def quantile(row):
    if row["amount_spent"]> quantile_95:
        return "VIP Customers"
    elif row["amount_spent"]< quantile_95 and row["amount_spent"]> quantile_75:
        return "Preffered Customers"
    else:
        return "Other Customers"


    
orders_grouped["Customer status"]= orders_grouped.apply(quantile, axis = 1)

display(orders_grouped)
        
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

merged_df = pd.merge(orders, orders_grouped["Customer status"], how = "left", on =["CustomerID"])

display(merged_df)
```

```python
df = merged_df[merged_df["Customer status"] == "VIP Customers"]

df.groupby(["Country"]).agg({"CustomerID":'count'}).sort_values(by="CustomerID", ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here

df = merged_df[~(merged_df["Customer status"] == "Other Customers")]

df.groupby(["Country","Customer status"]).agg({"CustomerID":'count'}).sort_values(by="CustomerID", ascending=False)
```
